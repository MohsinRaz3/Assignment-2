```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
slices=[17,12,8,8,34,21]
device_names=['Huawie','Apple','Xiaomi','Oppo','Others','Samsung']
```

```python

plt.pie(slices,
        labels=device_names
        ,radius=1.1,
        autopct='%0.0f%%',
        shadow=True,
        startangle=60),
        
plt.xlabel("Market share, Q1 2019")
plt.show()
```
